Taken a real hypothesis from our dataset:

H: Average purchase amount of males = average purchase amount of females
H: Average purchase amount of males  average purchase amount of females

This is perfect for a two-sample t-test, because youre comparing means between two groups.

```python
import scipy.stats as stats
```

```python
from scipy.stats import ttest_ind
male_purchase=df[df['Gender']=='Male']['Purchase Amount (USD)']
Female_purchase=df[df['Gender']=='Female']['Purchase Amount (USD)']                                  
```

t_stat (T-statistic): tells you how far apart the two group means are relative to their spread. Bigger absolute value  bigger difference.

p_value: tells you the probability that the difference happened by chance. Small p (<0.05)  difference is statistically significant.

```python
t_stat,p_value=ttest_ind(male_purchase,Female_purchase,equal_var=False)
print(f"T-statistic: {t_stat:.2f}, P-value: {p_value:.4f}")
```

P-value < 0.05  Reject H  There is a significant difference in average purchase amount between genders.

P-value  0.05  Fail to reject H  No significant difference.

We fail to reject H

This means, theres no statistically significant difference in average purchase amount between males and females in your dataset.

```python
from scipy.stats import f_oneway
category_groups=[df[df['Category']== categories]['Purchase Amount (USD)'] for categories in df['Category'].unique()]
f_stat,p_value=f_oneway(*category_groups)
print(f"f-satistic:{f_stat}, P-value:{p_value}")
```

ANOVA is used to compare the means of 3 or more groups to see if at least one group is significantly different from the others.

How it works:

ANOVA calculates the variance within each group and variance between groups.

If the between-group variance is large compared to within-group variance, the group means are likely significantly different.

ANOVA Test on Category vs Purchase Amount (USD) step by step.

Null hypothesis (H): All categories have the same average purchase amount

Alternative hypothesis (H): At least one category has a different average purchase amount

We fail to reject H
This means, theres no statistically significant difference  At least one category has a different average purchase amount
